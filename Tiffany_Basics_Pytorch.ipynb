```python
import torch.nn as nn
softmax = nn.Softmax(dim=0)
logits = torch.tensor([1.0, 2.0, 3.0])
probs = softmax(logits) #softmax change it to probabilities
print(probs)
print(probs.sum())
```

```python
class simpleclassifier(nn.Module):
  def __init__(self):
    super().__init__() #calls the class
    self.linear = nn.Linear(4,3) #nn.linear(in_features, out_features)
    self.softmax = nn.Softmax(dim=1)

  def forward(self, x):
    logits = self.linear(x)
    return self.softmax(logits)

model = simpleclassifier()
x = torch.rand(5,4)
output = model(x)

print(output)
print(output.sum(dim=1))


```

```python
model = nn.Sequential(nn.Linear(6,10), nn.ReLU(), nn.Linear(10,4), nn.Softmax(dim=1))
x = torch.rand(8,6)
output = model(x)
print(output)
print(output.sum(dim=1))
```

```python
import torch
import torch.nn as nn

layer = nn.Linear(3,2)
x = torch.tensor([[1.0, 2.0, 3.0]])
out = layer(x) #calls forward(x), forward pass

manual = x @ layer.weight.T + layer.bias #layer transpose to 2x3 so you can matrix multiply 1x3

print("Layer output: ", out)
print("Manual output:", manual)
```

```python
layer_with_bias = nn.Linear(4,2, bias=True)
layer_without_bias = nn.Linear(4,2, bias=False)
print ("with bias ", list(layer_with_bias.parameters()))
print("without bias", list(layer_without_bias.parameters()))
```

```python
layer = nn.Linear(3,2, bias=False)
with torch.no_grad():
  layer.weight = nn.Parameter(torch.tensor([
      [1.0,2.0,3.0],
      [0.0,1.0,0.0]
      ]))
  x = torch.tensor([[5.0,6.0,7.0]])
  print(layer(x))
```
